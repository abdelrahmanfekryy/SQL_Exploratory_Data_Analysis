```python
%%sql
SELECT t3.league_name,
        SUM(t2.goal) total_goals,
        ROUND(SUM(t2.goal)*100/CAST(SUM(SUM(t2.goal)) OVER () AS float),2) "total_goals %",
        COUNT(DISTINCT t2.match_api_id) match_count,
        ROUND(COUNT(DISTINCT t2.match_api_id)*100/CAST(SUM(COUNT(DISTINCT t2.match_api_id)) OVER () AS float),2) "match_count %",
        ROUND(CAST(sum(t2.goal) AS float)/CAST(COUNT(DISTINCT t2.match_api_id) AS float),2) "avg_goals/match"
FROM Match t1
LEFT JOIN Match_Team t2
ON t2.match_api_id = t1.match_api_id
LEFT JOIN League t3
ON t3.league_id = t1.league_id
GROUP BY t3.league_name
ORDER BY 6 DESC;
```

## Goals Count for each League over seasons

```python
%%sql
SELECT t3.league_name "League_Name",
        SUM(case when t1.season = "2008/2009" then t2.goal end) as "2008/2009",
        SUM(case when t1.season = "2009/2010" then t2.goal end) as "2009/2010",
        SUM(case when t1.season = "2010/2011" then t2.goal end) as "2010/2011",
        SUM(case when t1.season = "2011/2012" then t2.goal end) as "2011/2012",
        SUM(case when t1.season = "2012/2013" then t2.goal end) as "2012/2013",
        SUM(case when t1.season = "2013/2014" then t2.goal end) as "2013/2014",
        SUM(case when t1.season = "2014/2015" then t2.goal end) as "2014/2015",
        SUM(case when t1.season = "2015/2016" then t2.goal end) as "2015/2016",
        SUM(t2.goal) "Total_Goals",
        ROUND(SUM(t2.goal)*100/CAST(SUM(SUM(t2.goal)) OVER () AS float),2) "Total_Goals_%"
FROM Match t1
LEFT JOIN Match_Team t2
ON t2.match_api_id = t1.match_api_id
LEFT JOIN League t3
ON t3.league_id = t1.league_id
GROUP BY 1
ORDER BY 10 DESC;
```

## Average Goals per Match for each league over seasons

```python
%%sql
with Table1 AS (SELECT t3.league_name "League_Name",
                       t2.match_api_id,
                       t1.season,
                       SUM(t2.goal) "goal"
                FROM Match t1
                LEFT JOIN Match_Team t2
                ON t2.match_api_id = t1.match_api_id
                LEFT JOIN League t3
                ON t3.league_id = t1.league_id
                GROUP BY 1,2)
SELECT League_Name,
       ROUND(AVG(case when season = "2008/2009" then goal end),2) as "2008/2009",
       ROUND(AVG(case when season = "2009/2010" then goal end),2) as "2009/2010",
       ROUND(AVG(case when season = "2010/2011" then goal end),2) as "2010/2011",
       ROUND(AVG(case when season = "2011/2012" then goal end),2) as "2011/2012",
       ROUND(AVG(case when season = "2012/2013" then goal end),2) as "2012/2013",
       ROUND(AVG(case when season = "2013/2014" then goal end),2) as "2013/2014",
       ROUND(AVG(case when season = "2014/2015" then goal end),2) as "2014/2015",
       ROUND(AVG(case when season = "2015/2016" then goal end),2) as "2015/2016",
       ROUND(CAST(SUM(goal) AS float)/CAST(COUNT(DISTINCT match_api_id) AS float),2) "Total goals/match"
FROM Table1
GROUP BY 1
ORDER BY 10 DESC;
```

# Which League had the most matches end as draw in each season?

```python
%%sql
SELECT t3.league_name,
        SUM(case when t2.team_status = "home" then t2.goal end) as home,
        SUM(case when t2.team_status = "away" then t2.goal end) as away
FROM Match t1
LEFT JOIN Match_Team t2
ON t2.match_api_id = t1.match_api_id
LEFT JOIN League t3
ON t3.league_id = t1.league_id
GROUP BY 1;
```

```python
%%sql
WITH table1 AS (SELECT t3.league_name,
                        t2.match_api_id,
                        MAX(case when t2.team_status = "home" then t2.goal end) as home,
                        MAX(case when t2.team_status = "away" then t2.goal end) as away
                FROM Match t1
                LEFT JOIN Match_Team t2
                ON t2.match_api_id = t1.match_api_id
                LEFT JOIN League t3
                ON t3.league_id = t1.league_id
                GROUP BY 1,2)
SELECT league_name,
        COUNT(case when home = away then 1 else NULL end) "# draws",
        COUNT(case when home > away then 1 else NULL end) "# home win",
        COUNT(case when home < away then 1 else NULL end) "# away win"
FROM table1
GROUP BY 1
ORDER by 2 DESC;
```

```python
%%sql
SELECT t3.league_name,
        t2.match_api_id,
        t1.season,
        MAX(case when t2.team_status = "home" then t2.goal end) as home,
        MAX(case when t2.team_status = "away" then t2.goal end) as away
FROM Match t1
LEFT JOIN Match_Team t2
ON t2.match_api_id = t1.match_api_id
LEFT JOIN League t3
ON t3.league_id = t1.league_id
GROUP BY 1,2
HAVING home = away
LIMIT 10
```

# no. of draws for each league over seasons

```python
%%sql
WITH table1 AS (SELECT t3.league_name,
                        t2.match_api_id,
                        t1.season,
                        MAX(case when t2.team_status = "home" then t2.goal end) as home,
                        MAX(case when t2.team_status = "away" then t2.goal end) as away
                FROM Match t1
                LEFT JOIN Match_Team t2
                ON t2.match_api_id = t1.match_api_id
                LEFT JOIN League t3
                ON t3.league_id = t1.league_id
                GROUP BY 1,2
                HAVING home = away
                )
SELECT league_name,
        COUNT(case when season = "2008/2009" then match_api_id end) as "2008/2009",
        COUNT(case when season = "2009/2010" then match_api_id end) as "2009/2010",
        COUNT(case when season = "2010/2011" then match_api_id end) as "2010/2011",
        COUNT(case when season = "2011/2012" then match_api_id end) as "2011/2012",
        COUNT(case when season = "2012/2013" then match_api_id end) as "2012/2013",
        COUNT(case when season = "2013/2014" then match_api_id end) as "2013/2014",
        COUNT(case when season = "2014/2015" then match_api_id end) as "2014/2015",
        COUNT(case when season = "2015/2016" then match_api_id end) as "2015/2016",
        COUNT(DISTINCT match_api_id) AS "Total Draws"
FROM table1
GROUP BY 1
ORDER BY 10 DESC;
```
